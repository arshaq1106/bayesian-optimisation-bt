```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
tickers = ["BYDDF", "TSLA"]
start_date = "2023-01-01"
end_date = "2025-10-31"

raw_data = yf.download(tickers, start = start_date, end = end_date)
```

```python
prices = raw_data["Close"].copy()
prices.columns = tickers
```

```python
prices = prices.dropna(how="any")
```

```python
train_start = "2023-01-01"
train_end = "2025-01-31"

test_start = "2025-02-01"
test_end = "2025-10-31"
```

```python
prices_train = prices.loc[train_start:train_end].copy()
prices_train.tail()
```

```python
prices_test = prices.loc[test_start:test_end].copy()
prices_test.tail()
```

```python
prices_train.to_excel("train_data.xlsx")
```

```python
prices_test.to_excel("test_data.xlsx")
```
